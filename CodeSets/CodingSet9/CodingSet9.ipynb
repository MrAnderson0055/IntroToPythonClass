```python
#Programming Exercise #1
#Write a program that creates a dictionary containing course numbers and the room numbers of the rooms where the courses meet. 
#The dictionary should have the following key value pairs:

#CS101:3004
#CS102:4501
#CS103:6755
#NT110:1244
#CM241:1411

#The program should also create a dictionary containing course numbers and the names of the instructors that teach each course. 
#The dictionary should have the following key-value pairs:

#CS101:Haynes
#CS102:Alvarado
#CS103:Rich
#NT110:Burke
#CM241:Lee

#The program should also create a dictionary containing course numbers and the meeting times of each course. 
#The dictionary should have the following key-value pairs:

#CS101 8:00 a.m.
#CS102 9:00 a.m.
#CS103 10:00 a.m.
#NT110 11:00 a.m.
#CM241 1:00 p.m.

#The program should let the user enter a course number, then it should display the courseâ€™s room number, instructor, and meeting time.
```

```python
## Going to start by creating a function for the information.
def main():
    rooms={'CS101':'3004','CS102':'4501','CS103':'6755','NT110':'1244','CM241':'1411'}

    teachers={'CS101':'Haynes','CS102':'Alvarado','CS103':'Rich','NT110':'Burke','CM241':'Lee'}

    start_time={'CS101':'8:00am', 'CS102':'9:00am','CS103':'10:00am','NT110':'11:00am','CM241':'1:00pm'}
#going to ask for input for the course they need to know about
    course=input('What course do you need more inforamtion about?')
    if course in rooms: #Making a list of information for student if their input matches
        print("\nCourse Information")
        print(f'Room:{rooms[course]}')
        print(f'Instructor:{teachers[course]}') 
        print(f'Start Time:{start_time[course]}')
    else:
        print('Thats not a correct class...')


if __name__== "__main__":
    main()
    
    
```
