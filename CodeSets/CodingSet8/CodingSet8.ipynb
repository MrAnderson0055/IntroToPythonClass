```python
#Programming Exercise #5
#Many companies use telephone numbers like 555-GET-FOOD so the number is easier for their customers to remember. 
#On a standard telephone, the alphabetic letters are mapped to numbers in the following fashion:
```

```python
# Going to start with assigning letters to numbers. Then taking input and display it back to the user.

def numbertoletter(number):
    dial={ 'A': '2', 'B': '2', 'C': '2', 'D': '3', 'E': '3', 'F': '3', 'G': '4', 'H': '4', 'I': '4',
        'J': '5', 'K': '5', 'L': '5',
        'M': '6', 'N': '6', 'O': '6',
        'P': '7', 'Q': '7', 'R': '7', 'S': '7',
        'T': '8', 'U': '8', 'V': '8',
        'W': '9', 'X': '9', 'Y': '9', 'Z': '9'}

    switch=""

    for ch in number.upper():
        if ch.isalpha():
            switch +=dial[ch]
        else:
            switch+=ch
    return switch

user=input("Please enter a phone number")
switched_number=numbertoletter(user)
print(switched_number)

    



    
```

```python
#Programming Exercise #8
#Write a program with a function that accepts a string as an argument and returns a copy of the string with the first character of each sentence capitalized. 
#For instance, if the argument is “hello. my name is Joe. what is your name?” the function should return the string  
#The program should let the user enter a string and then pass it to the function. The modified string should be displayed.
```

```python
# Going to start by making it so the string can be read for periods and the add an UPPER to whatever comes after.

def caps(sentance):  #trying to have it read whatever is passed and store it
    sent=sentance.split('. ')
    new_sent=[]
    for sents in sent:
        sents=sents.strip() ## pulling from the input
        if sents:
            sents=sents[0].upper() +sents[1:]
            new_sent.append(sents)
    return '. '.join(new_sent)


user=input("please write a paragraph")
captext=caps(user)
print(captext)






            
    


```
